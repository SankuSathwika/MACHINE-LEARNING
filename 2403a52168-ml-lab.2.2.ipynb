```python
import pandas as pd
df = pd.read_csv('/content/Salary_dataset.csv')
```

```python
x = df.YearsExperience
y = df.Salary
```

```python
import numpy as np
```

```python
x=np.array(x).reshape(-1,1)
y=np.array(y).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train.values.reshape(-1, 1), y_train.values.reshape(-1, 1))
```

```python
model.fit(x_train.values.reshape(-1, 1), y_train.values.reshape(-1, 1))
y_pred = model.predict(x_test.values.reshape(-1, 1))
```

```python
print("Actual Salary:",y_test[:5])
print("Predicted Salary:",y_pred[:5])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
rmse = np.sqrt(mse)
print("Root Mean Squared Error:",rmse)
```

```python
from sklearn.metrics import r2_score
```

```python
R2 = r2_score(y_test,y_pred)
print("R-squared value:", R2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.show()
```

```python

plt.scatter(y_test,y_test,label="Actual Salary")
plt.scatter(y_test,y_pred,label="Predicted Salary")
plt.xlabel("Actual Salary")
plt.ylabel("Salary Value")
plt.title("Actual vs Predicted Salary Comparison")
plt.legend()
plt.show()
```

```python
slop = model.coef_[0]
intercept = model.intercept_
print("Slope (m):",slop)
print("Intercept (c):",intercept)
```
